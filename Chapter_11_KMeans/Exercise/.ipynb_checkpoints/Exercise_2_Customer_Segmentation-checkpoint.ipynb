### Step 4: Modeling & Evaluation/ Analyze & Report


```python
# Trực quan hóa dữ liệu để có cái nhìn ban đầu về việc phân nhóm
# Income & Age
plt.figure(figsize=(8,8))
sns.scatterplot(data=data, x='income', y='age')
plt.show()
```

```python
# Income, Age, Gender
plt.figure(figsize=(8,8))
sns.scatterplot(data=data, x='income', y='age', hue='gender_Male')
plt.show()
```

```python
"""* Quan sát các sơ đồ ta thấy dữ liệu có thể phân chia thành 4 cụm.
* Áp dụng thuật toán K-means để phân cụm, có thể kiểm tra lại số cụm (k)
bằng phương pháp Elbow method """
```

### Phân cụm theo mm_income, mm_age

```python
X = data_new[["mm_income", "mm_age"]]
X.head(2)
```

```python
# chọn k
distortions = []  # chứa WSSE tương ứng
k_list = range(1, 10)

for k in k_list:
    kmeans_model = KMeans(n_clusters=k).fit(X)  # số cụm
    distortions.append(
        sum(
            np.min(cdist(X, kmeans_model.cluster_centers_, "euclidean"), axis=1),
        )
        / X.shape[0]
    )
```

```python
distortions
```

```python
# Trực quan elbow
plt.plot(k_list, distortions, "bx-")
plt.xlabel("k")
plt.ylabel("Distortion")
plt.title("The Elbow Method showing the optimal k")
plt.show()
```

```python
# Chọn k = 4 vì từ 4 lỗi giảm đi nhỏ và mịn
```

```python
# áp dụng k = 4
kmeans_model = KMeans(n_clusters=4).fit(X)
kmeans_model
```

```python
centroids = kmeans_model.cluster_centers_
labels = kmeans_model.labels_
print("Centroids in min-max scale:")
print(centroids)
print("Centroids in normal:")
normal_centroids = mmScaler.inverse_transform(centroids)  # giá trị thật ko scaler
print(normal_centroids)
print(labels)
```

### Report

```python
data["Group"] = pd.Series(labels)
data.head()
```

```python
data.Group.value_counts()
```

```python
"""
=> mỗi nhóm có số lượng khách hàng khá bằng nhau (phân bố khá đều)
"""
```

```python
i = 0
for cluster in normal_centroids:
    print("*** Centroid Cluster: " + str(i))
    print("Income:", round(cluster[0], 2))
    print("Age:", round(cluster[1]), 2)
    i = i + 1
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(data.income, data.age, c=data.Group, cmap="rainbow")
plt.scatter(normal_centroids[:, 0], normal_centroids[:, 1], marker="s", c="black")
plt.show()
```

```python
"""
=> Có thể chia khách hàng thành 4 nhóm
Nhóm 1: Tuổi <=40 và thu nhập <=75K
Nhóm 2: Tuổi >40 và thu nhập <=75K
Nhóm 3: Tuổi <=40 và thu nhập >75K
Nhóm 4: Tuổi >40 và thu nhập >75K

"""
```

### Phân cụm theo mm_income, mm_age, Gender

```python
X = data_new[["mm_income", "mm_age", "gender_Male"]]
```

```python
distortions = []  # chứa WSSE tương ứng
k_list = range(1, 11)

for k in k_list:
    kmeans_model = KMeans(n_clusters=k).fit(X)
    distortions.append(
        sum(
            np.min(cdist(X, kmeans_model.cluster_centers_, metric="euclidean"), axis=1)
            / X.shape[0]
        )
    )
```

```python
distortions
```

```python
# Trực quan elbow
plt.figure(figsize=(18, 10))
plt.plot(k_list, distortions, "bx-")
plt.xlabel("k")
plt.ylabel("Distortion")
plt.title("The Elbow Method showing the optimal k")
plt.show()
```

```python
# Áp dụng k = 2 / hoặc k = 4
# Ví dụ: chọn k = 4
kmeans_3 = KMeans(n_clusters=4)
kmeans_3.fit(X)
```

```python
centroids_3 = kmeans_3.cluster_centers_
labels_3 = kmeans_3.labels_
print("Centroids in min-max scale:")
print(centroids_3)
print(labels_3)
```
