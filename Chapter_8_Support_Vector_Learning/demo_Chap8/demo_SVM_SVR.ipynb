```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
df = pd.read_csv('data/svr_data.csv')
df.head()
```

```python
plt.scatter(df['x'], df['y'])
plt.show()
```

```python
model_linear = SVR(kernel='linear', C=100)
model_poly = SVR(kernel='poly', C=100, degree=3)
model_rbf = SVR(kernel='rbf', C=100, gamma = 0.1)
```

```python
X = df[['x']]
y = df['y']
```

```python
model_linear.fit(X, y)
model_poly.fit(X, y)
model_rbf.fit(X, y)
```

```python
yhat_test_linear = model_linear.predict(X)
yhat_test_poly = model_poly.predict(X)
yhat_test_rbf = model_rbf.predict(X)
```

```python
# trực quan hóa kết quả
plt.figure(figsize=(15, 5))
plt.subplot(1,3,1)
plt.scatter(X, y)
plt.plot(X, yhat_test_linear, color='red')
plt.title('Linear Kernel')

plt.subplot(1,3,2)
plt.scatter(X, y)
plt.plot(X, yhat_test_poly, color='blue')
plt.title('Polynomial Kernel')

plt.subplot(1,3,3)
plt.scatter(X, y)
plt.plot(X, yhat_test_rbf, color='green')
plt.title('RBF Kernel')
plt.show()
```
